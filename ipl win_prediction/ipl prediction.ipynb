```python
import numpy as np
import pandas as pd
```

```python
match=pd.read_csv('iplm.csv')
delivery=pd.read_csv('iplb.csv')
```

```python
total_score_df=delivery.groupby(['ID','innings']).sum()['total_run'].reset_index()
```

```python
total_score_df=total_score_df[total_score_df['innings']==1]
```

```python
match_df=match.merge(total_score_df[['ID','total_run']],left_on='ID',right_on='ID')
```

```python
match_df['Team1'].unique()
```

```python
teams=[
    'Rajasthan Royals', 'Royal Challengers Bangalore',
       'Sunrisers Hyderabad', 'Delhi Capitals', 'Chennai Super Kings',
       'Gujarat Titans', 'Lucknow Super Giants', 'Kolkata Knight Riders',
        'Mumbai Indians', 'Kings XI Punjab'
]
```

```python
match_df['Team1']=match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2']=match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['Team1']=match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2']=match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1']=match_df['Team1'].str.replace('Punjab Kings','Kings XI Punjab')
match_df['Team2']=match_df['Team2'].str.replace('Punjab Kings','Kings XI Punjab')

```

```python
match_df=match_df[match_df['Team1'].isin(teams)]
match_df=match_df[match_df['Team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df=match_df[['ID','City','WinningTeam','total_run','Team1','Team2']]
```

```python
delivery_df=match_df.merge(delivery,on='ID')
```

```python
delivery_df
```

```python
delivery_df.shape
```

```python
delivery_df
```

```python
delivery_df['current_score']=delivery_df.groupby('ID').cumsum()['total_run_y']
```

```python
delivery_df['run_left']=delivery_df['total_run_x'] - delivery_df['current_score']
```

```python
delivery_df['ball_left']=120-(delivery_df['overs']*6+delivery_df['ballnumber'])
```

```python
delivery_df
```

```python
wicket=delivery_df.groupby('ID').cumsum()['isWicketDelivery'].values
delivery_df['wicket']=10-wicket
delivery_df.head(10)
```

```python
delivery_df['crr']=(delivery_df['current_score']*6)/(120-delivery_df['ball_left'])
```

```python
delivery_df['rrr']=(delivery_df['run_left']*6)/delivery_df['ball_left']
```

```python
def result(row):
    return 1 if row['BattingTeam']==row['WinningTeam'] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
final_df=delivery_df[['BattingTeam','City','run_left','ball_left','wicket','total_run_x','crr','rrr','result']]
```

```python
delivery_df
```

```python
conditions=[delivery_df['BattingTeam']==delivery_df['Team1'],delivery_df['BattingTeam']==delivery_df['Team2']]
```

```python
choices=[delivery_df['Team2'],delivery_df['Team1']]
delivery_df['BowlingTeam']=np.select(conditions,choices)
```

```python
delivery_df
```

```python
final_df=delivery_df[['BattingTeam','BowlingTeam','City','run_left','ball_left','wicket','total_run_x','crr','rrr','result']]
```

```python
final_df['BowlingTeam']=final_df['BowlingTeam'].astype(str)
final_df
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[final_df['ball_left']!=0]
```

```python
from sklearn.model_selection import train_test_split
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=1)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([('trf',OneHotEncoder(sparse_output=False,drop='first'),['BattingTeam','BowlingTeam','City'])],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear'))])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
